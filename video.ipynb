```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
from collections import Counter

# Given predictions and labels
predictions = ['Swing-Bench', 'SkateBoarding-Front', 'Run-Side', 'Golf-Swing-Side', 'Riding-Horse',
               'Lifting', 'Run-Side', 'Walk-Front', 'Kicking-Front', 'Run-Side',
               'Run-Side', 'Walk-Front', 'Swing-Bench', 'Riding-Horse', 'Golf-Swing-Front',
               'Riding-Horse', 'SkateBoarding-Front', 'Kicking-Side', 'Swing-Bench', 'Walk-Front',
               'Lifting', 'Kicking-Front', 'Swing-Bench', 'Walk-Front', 'Walk-Front',
               'Golf-Swing-Front', 'Diving-Side']

labels = ['Lifting', 'Diving-Side', 'Walk-Front', 'Golf-Swing-Side', 'Lifting',
          'Walk-Front', 'Kicking-Side', 'Walk-Front', 'Kicking-Front', 'Run-Side',
          'Run-Side', 'Walk-Front', 'Swing-Bench', 'Riding-Horse', 'Diving-Side',
          'Riding-Horse', 'SkateBoarding-Front', 'Walk-Front', 'Kicking-Side', 'Lifting',
          'Run-Side', 'Swing-Bench', 'Swing-Bench', 'Walk-Front', 'Walk-Front',
          'Lifting', 'Walk-Front']

print(Counter(predictions))
print(len(set(predictions)), len(set(labels)))

# Compute confusion matrix and accuracy
cm = confusion_matrix(labels, predictions)
accuracy = accuracy_score(labels, predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")

d = list(labels)
d.extend(predictions)

# Plot confusion matrix using seaborn
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)  # Adjust font size for better readability
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(d),
            yticklabels=np.unique(d))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
